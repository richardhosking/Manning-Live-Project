```python
"""
Live Project Deliverable 1.2 
Loading and segregating the CNN news dataset
clean data also - part of next section
export as a list in pkl format (serialized)

"""
import string 
import re
import os
import pickle
#import pdb # For Debugging
```

```python
# Working directories 
workingDirectory = '/home/richard/Desktop/Python Live Project/test'
cleanDirectory = workingDirectory+'/cleaned/'

# Working list which will hold cleaned summaries and news 
masterList = []

# Make directory for cleaned data
os.chdir(workingDirectory)
if os.path.isdir(cleanDirectory):
    pass
else:
    os.mkdir('cleaned')
    
# Get a list of the input files - ignore directories
files = [f for f in os.listdir(workingDirectory) if os.path.isfile(f)]
```

```python
#%%debug
# Clean and save to a Master List 
for filename in files:
    with open(filename, 'r') as infile, open(cleanDirectory+"dict_clean.pkl", "w") as dictfile, open(cleanDirectory + filename + "_clean.txt", "w") as outfile:
        # List containing all the lines in the file
        Lines = infile.readlines() 
        # Flag to indicate whether next line is a highlight
        highlightFlag = False
        # Create a dictionary to hold the news and highlights
        # and temporary working lists 
        newsDictionary = {'name': filename, 'news': [], 'summaries': []}
        news = []
        summaries= []
        
        for line in Lines:
            #pdb.set_trace()
            # make all one case
            # remember to add new string 'lowerline = ' strings are immutable!
            lowerline = line.lower()
            
            # Remove 'cnn'
            no_cnn = re.sub('cnn','', lowerline)
             
            # remove punctuation with regular expression - preserve @
            remove_breaks = re.sub('\n', '', no_cnn)
            no_punctuation = re.sub(r'[^\w\s\@]','', remove_breaks)
            # remove whitespace from beginning and end of line
            lstripped = no_punctuation.lstrip()
            stripped = lstripped.rstrip()
            
            # Separate news from highlights/summaries 
            # Summary is preceded by @highlight on previous line             
            # If line contains '@highlight' set highlight flag and go to next line
            if '@highlight' in stripped:
                highlightFlag = True
                #Debugging
                #pdb.set_trace()
                # Dont write line
                continue
            # Otherwise write to dictionary and outfile as long as there is something in the line
            if stripped == '':
                continue
            elif highlightFlag == True:
                summaries.append(stripped)
                highlightFlag = False
            else:
                news.append(stripped)
                
        newsDictionary['news'] = news
        newsDictionary['summaries'] = summaries 
        masterList.append(newsDictionary)
        
        # Write data to clean file 
        for line in news:
            outfile.write(line)
            outfile.write('\n')
            
        outfile.write("Summaries: ")
        outfile.write('\n')
        for line in summaries:
            outfile.write(line)
            outfile.write('\n')
            

# save masterList in serialized format
with open('dictfile', 'wb') as fp:
    pickle.dump(masterList, fp)

# and test result     
with open ('dictfile', 'rb') as fp:
    itemlist = pickle.load(fp)


```
